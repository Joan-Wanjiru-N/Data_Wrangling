## Clean

#### Define
- Select all records using Immediately,As soon as possible in the Start Date and replace that with ASAP
- Select all columns using nondescriptive column headers e.g. Applicationp and replace that with the full name for instance application procedure.

#### Code

```python
# we make a copy of the data set so can have the original data set to refer to and clean the copy 
df_clean = df.copy()
```

>  Select all columns using nondescriptive column headers e.g. Applicationp and replace that with the full name for instance application procedure.

```python
df_clean.columns
```

```python
df_clean = df_clean.rename(columns = {"RequiredQual": "RequiredQualifications","ApplicationP": "ApplicationProcedure","AboutC":"AboutCompany", "JobRequirment": "JobRequirments" })
df_clean.columns
```

```python
# For consistency we convert characters in columns to lower
df_clean.rename(str.lower, axis = 'columns', inplace = True)
df_clean.columns
```

> Select all records using Immediately,As soon as possible in the Start Date and replace that with ASAP

```python
df_clean.startdate.value_counts()
```

```python
#create a list for items to be replaced
asap_list = ['Immediately', 'As soon as possible', 'Upon hiring',
             'Immediate', 'Immediate employment', 'As soon as possible.', 'Immediate job opportunity',
             '"Immediate employment, after passing the interview."',
             'ASAP preferred', 'Employment contract signature date',
             'Immediate employment opportunity', 'Immidiately', 'ASA',
             'Asap', '"The position is open immediately but has a flexible start date depending on the candidates earliest availability."',
             'Immediately upon agreement', '20 November 2014 or ASAP',
             'immediately', 'Immediatelly',
             '"Immediately upon selection or no later than November 15, 2009."',
             'Immediate job opening', 'Immediate hiring', 'Upon selection',
             'As soon as practical', 'Immadiate', 'As soon as posible',
             'Immediately with 2 months probation period',
             '12 November 2012 or ASAP', 'Immediate employment after passing the interview',
             'Immediately/ upon agreement', '01 September 2014 or ASAP',
             'Immediately or as per agreement', 'as soon as possible',
             'As soon as Possible', 'in the nearest future', 'immediate',
             '01 April 2014 or ASAP', 'Immidiatly', 'Urgent',
             'Immediate or earliest possible', 'Immediate hire',
             'Earliest  possible', 'ASAP with 3 months probation period.',
             'Immediate employment opportunity.', 'Immediate employment.',
             'Immidietly', 'Imminent', 'September 2014 or ASAP', 'Imediately']

# use for loop to replace all items in asap_list wist ASAP
for i in asap_list: 
    df_clean.startdate.replace(i ,"ASAP" , inplace=True)
    
#Test
df_clean.startdate.value_counts()
```

#### Test

```python
# since cant see all the values while using value_counts() let use assert to test all the records in asap_list exists in df_clean startdate

for i in asap_list:
    assert i not in df_clean.startdate.values

# is results of assert are true then nothing ir returned else if false an assertion error is returned 
```

## Analysis & Visualization
 
<br/>

> What pencent rate are the jobs that require urgent reporting?


```python
#get the count of reccords that require urgent reporting and assign to avarible
asap_counts = df_clean.startdate.value_counts()['ASAP']

asap_counts
```

```python
#get all records in startdate that are non null
startdate_counts = df_clean.startdate.count()

startdate_counts
```

```python
#percent Rate for jobs requiring immediate reporting
Rate = round((asap_counts / startdate_counts) * 100, 2)

Rate
```

```python
#lets crate a visual for the count of items in startdate, and assign other records except ASAP to blank labels
labels = np.full(len(df_clean['startdate'].value_counts()), "", dtype = object)
labels[0] = "ASAP"
df_clean['startdate'].value_counts().plot(kind = 'pie', figsize = (8,8), labels = labels);
```

```python
# Let compare start date record value_counts with that of the data before cleaning
labels = np.full(len(df['StartDate'].value_counts()), "", dtype = object)
labels[0] = "ASAP"
df['StartDate'].value_counts().plot(kind = 'pie', figsize = (8,8), labels = labels);
```
